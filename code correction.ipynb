```python
def otlier_treat(df,column):
    
    for i in df[column]:
        Q1 = np.quantile(df[column], .25) 
        Q3 = np.quantile(df[column], .75) 
        Q2 = np.quantile(df[column], .50)
        
        IQR = Q3-Q1
        lower = Q1 -(1.5 * IQR) 
        upper = Q3 +(1.5 * IQR)
        
        if i > upper or i < lower:
            df[column]=df[column].replace(i, np.median(df[column]))
```
